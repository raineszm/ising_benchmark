```python
N = 25
#Spins are 1 or -1
spins = ones((N,N), int)
#Using reduced units
#These are just here for book-keeping
J = 1


def delta_E(s, i, j):    
    return 2*J*s[i,j]*(s[remainder(i+1,N),j] + s[i,remainder(j+1,N)] + s[remainder(i-1,N),j] + s[i, remainder(j-1,N)])
    
def spin_flip(s, i, j):
    s[i,j] = - s[i,j]

def accepted(dE, beta):
    P = exp(-beta*dE)
    return rand() < P

def metropolis_step(s, beta):
    i,j = randint(0,N,2)
    dE = delta_E(s,i,j)
    if dE < 0 or accepted(dE, beta):
        spin_flip(s,i,j)

def evolve(s, n, beta):
    for i in range(n):
        metropolis_step(s,beta)


def time_average(s, n, f, beta=1):
    values = empty(n, int)
    for i in range(n):
        metropolis_step(s, beta)
        values[i] = f(s)
        
    return values.mean()
    
def magnetization(s):
    return sum(s)

def energy(s):
    return -J*sum(s*(roll(s,1,0) + roll(s,-1,0) + roll(s,1,1) + roll(s,-1,1)))
            
@vectorize
def ensemble_av(beta, f, n_evolve=1000, n_average=100):
    s = spins.copy()
    evolve(s, n_evolve, beta)
    return time_average(s, n_average, f, beta)
```

```python
z=linspace(0.001,10,200)
ms=ensemble_av(1/z,magnetization,500,100)
```

```python
es = ensemble_av(1/z,energy,500,100)
cs = diff(es)/diff(z)
```

```python
plot(z,ms)
xlabel("$k_B T/J$")
ylabel("$M/\mu_B$")
```

```python
plot(z,es)
xlabel("$k_B T/J$")
ylabel("$E/J$")
title("Total Energy vs Temperature")
```

```python
plot(z[:-1],cs)
xlabel("$k_B T/J$")
ylabel("C_v/k_B")
title("Specific Heat vs Temperature")
```
