```python
data = load("data.npz")
T=data["T"]
M=data["M"]
U=data["U"]
N=float(data["N"])
Nsites = N**2
```

```python
T_c = 2/log(1+sqrt(2))
```

```python
plot(T,M/Nsites,'bx', label='Simulation')
plot(T, (1-sinh(2./T)**(-4.))**(1/8.), 'k',label='Exact')
xlabel("$\\frac{k_B T}{J}$", size=20)
ylabel("$\\frac{M}{N}$",size=20)
title("Magnetization vs Temperature")
legend()
axvline(x=T_c,color='k',linestyle='--')
annotate("$T_c$", (T_c-0.1,0), xytext=(1,0.6), arrowprops={'arrowstyle':'->'},size="32")
gcf().set_size_inches((12,8))
gcf().set_dpi(200)
```

```python
plot(T, U/Nsites,".")
xlabel("$\\frac{k_b T}{J}$",size=20)
ylabel("$\\frac{U}{NJ}$",size=20)
title("Energy vs Temperature")
axvline(x=T_c,color='k',linestyle='--')
annotate("$T_c$", (T_c-0.1,-1.4), (1,-1), arrowprops={"arrowstyle":"->"},size=32)
gcf().set_size_inches((12,8))
gcf().set_dpi(200)
```

```python
fdata = load("data-fine.npz")
T1, T2, T3 = (fdata[x] for x in ["T1", "T2", "T3"])
U1, U2, U3 = (fdata[x] for x in ["U1", "U2", "U3"])
def simple_deriv(z, f, n):
    return (roll(f,n) - roll(f,-n))/(roll(z,n) - roll(z,-n))

n1=5
n2=15
n3=10
ix1 = find(T < T2[0])
ix2 = find(T > T1[-1])[:T2.size]
ix3 = find(T > T2[-1])
dU1 = simple_deriv(T,U/Nsites,n1)[ix1]
dU2 = simple_deriv(T,U/Nsites,n2)[ix2]
dU3 = simple_deriv(T,U/Nsites,n3)[ix3]
dU = concatenate((dU1[n1:], dU2, dU3[:-n3]))
Tprime =T[n1:-n3]
plot(Tprime,dU, 'b.')
xlabel("$\\frac{k_B T}{J}$",size=20)
ylabel("$\\frac{C}{Nk_B}$",size=20)
axvline(x=T_c,color='k',linestyle='--')
annotate("$T_c$", (T_c-0.1,0.1), (1,1), arrowprops={"arrowstyle":"->"},size=24)
title("Specific Heat vs Temperature")
gcf().set_size_inches((12,8))
gcf().set_dpi(200)
```
